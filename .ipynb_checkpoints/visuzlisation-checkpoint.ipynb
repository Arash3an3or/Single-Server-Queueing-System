```python
mu = [10, 100, 3, 1000]
N = [20, 5, 1000, 10]
```

```python
def draw(message, func_from_class, loops=10, iterations=1000, \
         mu=[10, 100, 3, 1000], N=[20, 5, 1000, 10], \
        ):
    plt.figure(figsize=(15, 10))
    plt.suptitle(message)
    for i in range(len(mu)):
        lambd = range(mu[i]-(mu[i]//2+1), mu[i]+mu[i]//2+2)
        expect = []
        for l in lambd:
            expect_mean = []
            for loop in range(loops):
                expect_mean.append(func_from_class(iterations, l, mu[i], N[i]))
            expect.append(np.mean(expect_mean))
        plt.subplot(221+i)    
        plt.plot(lambd, expect, 'o-')
        plt.title('mu = {}, N = {}'.format(mu[i], N[i]))
        plt.xlabel('lambda')
        plt.ylabel('average number of items')
        plt.grid()
    plt.show()
```

```python
draw('2. Вероятность отказа в сист. в зависимости от lambda', \
     v2.prob_of_non_serving, iterations=100)
```

```python
draw('3. Среднее кол-во требований в сист. в зависимости от lambda', v2.mean_sojourn_time)
# loops=10, iterations=1000, \
# mu=[10, 100, 3, 1000], N=[20, 5, 1000, 10]
```
